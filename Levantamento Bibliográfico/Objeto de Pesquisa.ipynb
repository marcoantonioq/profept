```python
import pandas as pd

# Função interna para corrigir os desalinhamentos de cada linha do DataFrame
def corrigir_linha_df(row):
    # Garante que estamos manipulando uma cópia da linha
    nova_row = row.copy()

    # 1. Tratamento para as linhas que possuem o delimitador "||http" na coluna 'Link de acesso'
    val_link = str(nova_row['Link de acesso'])
    if "||http" in val_link:
        partes = val_link.split("||")
        texto_referencia = partes[0]
        url_link = partes[1]

        # Atribui cada valor à sua respectiva coluna correta
        nova_row['Link de acesso'] = url_link
        nova_row['Referência Bibliográfica'] = texto_referencia

    # 2. Tratamento para as colunas onde a referência bibliográfica veio colada no Link de acesso (sem "||")
    elif val_link.strip().startswith(("PINTO,", "ALVES,", "VIDAL,")):
        # Identifica se há um link de repositório HTTP escondido dentro do texto
        if "http" in val_link:
            pos_http = val_link.find("http")
            texto_referencia = val_link[:pos_http].strip().rstrip("||")
            url_link = val_link[pos_http:].strip()

            nova_row['Link de acesso'] = url_link
            nova_row['Referência Bibliográfica'] = texto_referencia

    # 3. Tratamento para a última linha truncada (Kennyson Carvalho)
    # Se o resumo ou outra coluna essencial estiver com texto inacabado, removemos as aspas soltas
    for col in nova_row.index:
        if isinstance(nova_row[col], str) and nova_row[col].endswith('"'):
            nova_row[col] = nova_row[col].rstrip('"')

    return nova_row

# Aplicando a correção linha por linha em todo o DataFrame carregado
df = df.apply(corrigir_linha_df, axis=1)

# Substituindo eventuais valores vazios/nulos pelo padrão do seu arquivo
df = df.fillna("Não informado pela instituição")

print(f"Planilha corrigida com sucesso! {df.shape[0]} linhas e {df.shape[1]} colunas totalmente alinhadas.")
```
